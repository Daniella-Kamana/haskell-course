## Expressions If-then-else

Souvent, dans votre code, vous devez faire un choix. Il existe plusieurs façons d'exprimer des conditions. En Haskell, nous utilisons le plus souvent des expressions **if-then-else** :

```haskell
if <Condition> 
 then <Expesssion1>
 else <Expesssion2>
```

Où `Condition` est une expression logique qui renvoie `True` ou `False`, `Expression1` est l'expression utilisée si `Condition` est `True`, et `Expression2` est l'expression utilisée si `Condition` est `Faux`. La fonction `checkLocalHost` ci-dessous vérifie si l'argument est localhost ou non et le signale à l'utilisateur.

````haskell
checkLocalhost :: String -> String
checkLocalhost ip =
 -- Vrai ou faux ?
 if ip == "127.0.0.1"
 -- Lorsque la condition est vraie, la réponse est
 then "It's localhost!"
 -- Sinon la condition est False et la réponse est
 else "Non, ce n'est pas localhost."

checkLocalhost "127.0.0.1"
````

La fonction `checkLocalhost` est appliquée à un seul argument de type `String` et renvoie une autre valeur de type `String`. L'argument est une chaîne `ip` contenant l'adresse IP, et la fonction vérifie si la chaîne est égale à `"127.0.0.1"`. Si la vérification est réussie, la fonction renvoie « It's localhost! », sinon elle renvoie « No, it's not localhost. » 

<div class="alert alert-block alert-info">
 Alors que dans les langages de programmation impératifs, le <code>else</code> n'est pas obligatoire, en Haskell, il l'est ! C'est parce que, dans Haskell, chaque fonction doit renvoyer une valeur. Nous sommes donc obligés de fournir un résultat du même type pour les cas <code>then</code> et <code>else</code>. 
</div>

## Guards

Imaginons maintenant que nous souhaitons effectuer une vérification plus complexe. Comme vérifier si l'anniversaire de cette année a une signification particulière. Nous pourrions utiliser des instructions if-else imbriquées comme celle-ci :

````haskell
specialBirthday :: Int -> [Char]
specialBirthday age =
  if age == 1
    then "First birthday!"
    else
      if age == 18
        then "Tu es un adulte!"
        else
          if age == 60
            then "Enfin, je peux arrêter de me soucier du nouveau jargon!"
            else "Rien de spécial"
````

C'est juste un gâchis ! Trop compliqué à lire et à écrire. Heureusement, nous avons des gardes

Les gardes fonctionnent de la même manière que les instructions if-else, mais vous pouvez avoir plusieurs conditions:

```haskell
func arg
  | <Condition1> = <Result1>
  | <Condition2> = <Result2>
  | <Condition3> = <Result3> 
  ...
```

Nous utilisons le symbole `|` pour indiquer le début de chaque garde.

<div class="alert alert-block alert-info">
    Notez qu'il n'y a pas de signe <code>=</code>après les arguments <code>func</code> C'est un piège courant lors de l'écriture de gardes. N'ajoutez pas ce <code>=</code>!
</div>

Avec les gardes, nous pouvons écrire la fonction `specialBirthday`comme ceci:

````haskell
specialBirthday :: Int -> [Char]
specialBirthday age
  | age == 1 = "Premier anniversaire!"
  | age == 18 = "Tu es un adulte!"
  | age == 60 = "Enfin, je peux arrêter de me soucier du nouveau jargon!"
  | True = "Rien de spécial"
````

Ce dernier `True` est là pour être une condition fourre-tout. Une condition qui s'évalue toujours à `True` parce qu'elle est littéralement `True`.

Ce modèle d'ajout d'un dernier `True` dans la dernière garde est si courant que Haskell est livré avec une variable appelée `otherwise` qui est égale à `True` (`otherwise = True`) pour rendre la garde encore plus lisible:

````haskell
specialBirthday :: Int -> [Char]
specialBirthday age
  | age == 1 = "Premier anniversaire!"
  | age == 18 = "Tu es un adulte!"
  | age == 60 = "Enfin, je peux arrêter de me soucier du nouveau jargon!"
  | otherwise = "Rien de spécial"

specialBirthday 60
````

Vous pouvez maintenant facilement comprendre ce que fait cette expression d'un simple coup d'œil!

OK, c'est tout pour les évaluations conditionnelles. Voyons maintenant comment nous pouvons améliorer notre jeu de syntaxe de fonction avec `let` et `where`!

## `let` et `where`

Nous utilisons `let` et `where` pour stocker les résultats des calculs intermédiaires et lier les variables locales.

 Let's start with `let`!

###  expressions `let`

`let` peut lier des expressions à des variables locales de la manière suivante:

```haskell
func arg =
    let <BIND_1> 
        <BIND_2> 
    in  <EXPR qui utilise BIND_1 and/or BIND_2>
```

Where `<BIND_X>` sont des liaisons locales accessibles dans toute l'expression `let`..

Maintenant, créons une fonction qui prend deux températures, une en Celsius et une en Fahrenheit, et renvoie la plus chaude mais en Kelvin. Ce sont de nombreuses conversions, n'est-ce pas ?

Pour passer de Fahrenheit à Celsius, nous devons d'abord soustraire 32 puis multiplier par 5/9, comme ceci:

 $tC = (tF - 32) * 5/9$

Pour passer de Celsius à Kelvin, il suffit d'ajouter 273,16 comme ceci:

$tK = tC + 273.16$

Donc, si nous voulons créer **une seule fonction** qui fait tout cela, nous pouvons créer quelque chose comme ceci:

````haskell
hotterInKelvin :: Double -> Double -> Double
hotterInKelvin c f = if c > (f - 32) * 5 / 9 then c + 273.16 else ((f - 32) * 5 / 9) + 273.16

hotterInKelvin 40 100
````

Ça marche, mais c'est du manuel Je-n'ai-aucune-idée-de-ce-que-je-voulais-faire-avec-ce-code-d'il-y-a-deux-semaines.

Une meilleure approche consiste à utiliser des liaisons `let` pour les expressions intermédiaires et à écrire l'expression qui rassemble tout dans la partie `in`:

````haskell
hotterInKelvin' :: Double -> Double -> Double
hotterInKelvin' c f =
  let fToC t = (t - 32) * 5 / 9
      cToK t = t + 273.16
      fToK t = cToK (fToC t)
   in if c > fToC f then cToK c else fToK f

hotterInKelvin' 40 100
````

Maintenant, notre code est beaucoup plus lisible et ne contient plus toutes ces expressions répétées!

Mais attendez, il y a plus ! Nous pouvons également utiliser la construction `where`!

### `where`

Nous pouvons utiliser `where` pour lier des valeurs à des variables de la manière suivante:

```haskell
func arg = <EXP that uses BIND_1 and/or BIND_2>
    where <BIND_1>
          <BIND_2>
```

Ainsi, la même fonction `hotterInKelvin` que précédemment peut être exprimée avec `where` comme ceci:

Où `<BIND_X>` sont des liaisons accessibles dans tout le corps de la fonction.

````haskell
hotterInKelvin'' :: Double -> Double -> Double
hotterInKelvin'' c f = if c > fToC f then cToK c else fToK f
  where
    fToC t = (t - 32) * 5 / 9
    cToK t = t + 273.16
    fToK t = cToK (fToC t)

hotterInKelvin'' 40 100
````

Ok, ils semblent tous les deux faire la même chose. Alors, pourquoi s'embêter à avoir les deux ? Ne pourrions-nous pas simplement choisir d'en utiliser un?

Eh bien, il existe de nombreux cas où ils sont interchangeables. Dans ces cas-là, vous pouvez choisir celui que vous préférez. Mais elles ont aussi leurs limites et leurs points forts.

### Dois-je utiliser `let` ou `where`?

`Les expressions `let` sont pratiques lorsque nous voulons diviser des expressions complexes en blocs de construction plus petits que vous combinez dans une expression finale.

Par exemple, imaginez que vous souhaitiez calculer le volume d'une maison. Nous pourrions simplifier le problème comme ceci:

Une maison est un cube avec une pyramide au sommet (le toit). Donc, pour trouver son volume, nous devons calculer le volume du cube et le volume de la pyramide et les additionner:

````haskell
houseV side roofH = let cubeV = side ^ 3
                        pyramidV = (side ^ 2) * roofH / 3
                    in  cubeV + pyramidV
                    
houseV 3 1
````

Nous créons les blocs de construction `cubeV` et `pyramidV` à l'intérieur du bloc `let`, puis nous les utilisons à l'intérieur de l'expression `in`.

Outre la clarté de la syntaxe, un autre avantage est que si l'expression finale devient plus compliquée par la suite (par exemple, nous ajoutons une cheminée à la maison), nous devons simplement ajouter une autre liaison et l'utiliser dans l'expression finale:

````haskell
houseV side roofH = let cubeV = side ^ 3
                        pyramidV = (side ^ 2) * roofH / 3
                        chimneyV = (0.5 ^ 2) * roofH
                    in  cubeV + pyramidV + chimneyV
                    
houseV 3 1
````

D'un autre côté, les expressions `where` sont pratiques chaque fois que nous voulons étendre les liaisons à plusieurs équations protégées.

Parce que nous ne pouvons pas accéder aux liaisons `let` sur toutes les protections, mais avec les liaisons `where`, nous le pouvons !!
Par exemple:

````haskell
analyzeCylinder :: Float -> Float -> String
analyzeCylinder diamètre hauteur
       | volume < 10 = "Le cylindre est un verre."
       | volume < 100 = "Le cylindre est un seau."
       | volume < 1000 = "Le cylindre est un réservoir."
       | otherwise = "Mais c'est quoi ce truc énorme?!"
    where
        volume = pi * diameter^2 * height / 4

analyzeCylinder 15 6
````

Comme vous pouvez le voir, nous avons défini la liaison `volume` à l'intérieur du bloc `where`, puis nous y avons accédé sur chaque expression protégée!

Et enfin, la principale différence entre les deux est que les liaisons `where` sont des déclarations liées à une construction syntaxique environnante. Cela signifie qu'ils ne peuvent être utilisés qu'à des endroits spécifiques (comme à l'intérieur d'un corps de fonction). Mais `let` introduit une expression, il peut donc être utilisé partout où une expression peut être utilisée. Par exemple:

````haskell
-- Secondes dans une journée
24 * (let seconds = 60 in seconds * 60) 

-- Le volume d'un prisme rectangulaire (on peut séparer les expressions par des points-virgules pour les avoir sur la même ligne)
let s1 = 10; s2 = 20; s3 = 30; in s1*s2*s3 
````

Dans tous les cas où vous pourriez utiliser l'un ou l'autre, choisissez celui qui convient à la situation ou à votre style. Il faut un peu de pratique pour choisir correctement lequel utiliser, et il y a aussi la préférence du programmeur. Ne vous en souciez donc pas trop. 

### Choses à garder à l'esprit

Les expressions définies avec `where` ne sont pas accessibles en dehors du corps de cette fonction.

````haskell
fToK t = 273.16 + fToC t
    where fToC t = (t - 32) * 5 / 9
    
fToC 60
````

L'expression introduite dans une expression `let` n'existe que dans cette expression `let`.

Par exemple, cette fonction prend votre nom et votre prénom et renvoie vos initiales:

````haskell
initials :: String -> String -> String
initials name lastName = if name == "" || lastName == ""
                         then "Comment était ton nom déjà?"
                         else let x = head name
                                  y = head lastName
                              in [x] ++ "." ++ [y] ++ "."

initials "Richard" "Feynman"
````

Les expressions `x` et `y` ne sont disponibles qu'à l'intérieur de cette expression `let`. Si vous essayez de les utiliser dans `if` ou `then`, ils seraient hors de portée et la compilation ne se ferait pas.
